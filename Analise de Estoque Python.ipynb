```python
# Atribiundo a base de dados à uma varíavel
dados=pd.read_csv("Análise de Estoque.csv",sep=";")
display(dados)
```

```python
#Apresentando estrutura da base de dados
display(dados.info())
```

```python
#Corrigindo o problema na base de dados
display(dados.columns)
```

```python
dados["PREÇO UNITÁRIO"] = dados ["PREÇO UNITÁRIO"].str.replace(",",".")
```

```python
dados["PREÇO UNITÁRIO"] = pd.to_numeric(dados["PREÇO UNITÁRIO"],errors ="coerce")
```

```python
display(dados)
```

```python
import plotly_express as px
```

```python
# Estoque por Loja
px.histogram(dados,x="ID LOJA",y="MOVIMENTAÇÃO",text_auto=True,color="ID LOJA")
```

```python
#Estoque por categoria
px.histogram(dados,x="ID LOJA",y="MOVIMENTAÇÃO",text_auto=True,color="CATEGORIA")
```

```python
#Estoque por Categoria e Loja
px.histogram(dados,x=["ID LOJA","CATEGORIA"],y="MOVIMENTAÇÃO",text_auto=True,color="CATEGORIA")
```

```python
#Quantidades de Entradas Saídas e Devoluções por Loja
px.histogram(dados,x="ID LOJA",y="MOVIMENTAÇÃO",text_auto=True,color="TIPO")
```

```python
#Quantidades de Entradas Saídas e Devoluções total
esd = px.histogram(dados,x="TIPO",y="MOVIMENTAÇÃO",text_auto=True,color="TIPO")
```

```python
esd.write_html("Ánalise de esd totais.html")
```
